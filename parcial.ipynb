### eliminando multicolinealidad con VIF

```python
import pandas as pd
from statsmodels.stats.outliers_influence import variance_inflation_factor

predictive_variables = wind_speed_data.drop(["Hora (UTC)", "Velocidad del Viento (m/s)"], axis = 1)

vif_data = pd.DataFrame()
vif_data["Variable"] = predictive_variables.columns
vif_data["VIF"] = [variance_inflation_factor(predictive_variables.values, i) for i in range(predictive_variables.shape[1])]

# Display the result
print(vif_data)
```

después de eliminar variables iterativamente, se llegó a este conjunto de variables que deacuerdo al variance inflation factor no presentan un VIF mayor que 5, por lo cuál no hay multicolinealidad presente

```python
predictive_variables_VIF = predictive_variables.drop([
    "Humedad Relativa Máxima (AUT) (%)",
    "Humedad Relativa Mínima (AUT) (%)",
    "Temperatura Máxima (AUT) (°C)",
    "Presión Atmosférica Máxima Anterior (AUT) (mB)",
    "Presión Atmosférica Mínima Anterior (AUT) (mB)",
    "Presión Atmosférica Horaria (mB)",
    "Temperatura Mínima (AUT) (°C)"
    ], axis = 1)

vif_data = pd.DataFrame()
vif_data["Variable"] = predictive_variables_VIF.columns
vif_data["VIF"] = [variance_inflation_factor(predictive_variables_VIF.values, i) for i in range(predictive_variables_VIF.shape[1])]

# Display the result
print(vif_data)
```

### modelos

```python
X = wind_speed_data[["Dirección del Viento (grados)",
                     "Humedad Relativa del Aire (%)",
                     "Precipitación Total Horaria (mm)",
                     "Rajada Máxima del Viento (m/s)"]]

y = wind_speed_data[["Velocidad del Viento (m/s)"]]
```

### KNN

```python
from sklearn.preprocessing import StandardScaler

# Assuming your data is in a variable called 'X'
# Create a StandardScaler object
scaler = StandardScaler()

# Fit the scaler to your data and transform it
X_scaled = scaler.fit_transform(X)

# 'X_scaled' now contains your standardized data
```

```python
from sklearn.neighbors import KNeighborsRegressor
from sklearn.metrics import mean_squared_error
from sklearn.preprocessing import StandardScaler
import numpy as np

# grilla de parametros a probar en knn regressor
neighbors_grid = [1,3,5,7,10,15,20,30,50,100]

# aqui range indica el numero de folds que se va a crear,
# se limita porque hay demasiados folds de 24 horas
num_folds = range(10)

# diccionario para almacenar los scores
scores_dict_knn = {}

final_index = 24

scaler = StandardScaler()

for n_neighbors in neighbors_grid:
    
    # lista para guardar los rmse en cada fold
    rmse_in_fold = []
    
    for fold in num_folds:

        X_train = X.iloc[0:final_index]
        X_test = X.iloc[final_index:final_index + 24]
        y_train = y.iloc[0:final_index]
        y_test = y.iloc[final_index:final_index + 24]
        
        final_index += 24
        
        X_train = scaler.fit_transform(X_train)
        X_test = scaler.fit_transform(X_test)

        ######## model training
        reg = KNeighborsRegressor(n_neighbors = n_neighbors)
        reg.fit(X_train, y_train)

        ######## calculating RMSE
        y_pred = reg.predict(X_test)

        residuals = y_test - y_pred

        mse = mean_squared_error(y_test, y_pred)
        rmse = np.sqrt(mse)
        
        rmse_in_fold.append(rmse)
        
    rmse_mean = np.mean(rmse_in_fold)
    
    scores_dict_knn[n_neighbors] = rmse_mean   
```

Nótese que en este diccionario, cada "key" corresponde al número de vecinos utilizado, y cada "value" corresponde al RMSE correspondiente

### Ridge

```python
from sklearn.linear_model import Ridge

# grilla de parametros a probar en knn regressor
alpha_grid = [0.1, 1.0, 10.0]

# aqui range indica el numero de folds que se va a crear,
# se limita porque hay demasiados folds de 24 horas
num_folds = range(10)

# diccionario para almacenar los scores
scores_dict_ridge = {}

final_index = 24

scaler = StandardScaler()

for alpha in alpha_grid:
    
    # lista para guardar los rmse en cada fold
    rmse_in_fold = []
    
    for fold in num_folds:

        X_train = X.iloc[0:final_index]
        X_test = X.iloc[final_index:final_index + 24]
        y_train = y.iloc[0:final_index]
        y_test = y.iloc[final_index:final_index + 24]
        
        final_index += 24
        
        X_train = scaler.fit_transform(X_train)
        X_test = scaler.fit_transform(X_test)

        ######## model training
        reg = Ridge(alpha = alpha)
        reg.fit(X_train, y_train)

        ######## calculating RMSE
        y_pred = reg.predict(X_test)

        residuals = y_test - y_pred

        mse = mean_squared_error(y_test, y_pred)
        rmse = np.sqrt(mse)
        
        rmse_in_fold.append(rmse)
        
    rmse_mean = np.mean(rmse_in_fold)
    
    scores_dict_ridge[alpha] = rmse_mean   
```
